# 提取特征

## 提取并生成hour、year变量

```python
all_data['year']=all_data.index.year
all_data['hour']=all_data.index.hour
all_data.columns
```

## 离散型变量转换

```python
all_data=pd.get_dummies(all_data,columns=['hour'],prefix=['hour'],prefix_sep=' ',drop_first=True)
all_data=pd.get_dummies(all_data,columns=['season'],prefix=['season'],prefix_sep=' ',drop_first=True)
all_data=pd.get_dummies(all_data,columns=['weather'],prefix=['weather'],prefix_sep=' ',drop_first=True)
all_data=pd.get_dummies(all_data,columns=['year'],prefix=['year'],prefix_sep=' ',drop_first=True)
```

## 删除不必要的变量

```python
all_data.head()
```

# 建模预测

## 从处理后的数据集中提取训练集和测试集

```python
new_train_data=all_data.iloc[:10886]
cols=['holiday', 'atemp', 'workingday','hour 1', 'hour 2', 'hour 3', 'hour 4', 'hour 5', 'hour 6', 'hour 7', 'hour 8', 'hour 9', 'hour 10', 'hour 11','hour 12', 'hour 13', 'hour 14', 'hour 15', 'hour 16', 'hour 17', 'hour 18', 'hour 19', 'hour 20', 'hour 21','hour 22', 'hour 23','year 2012', 'season 2', 'season 3', 'season 4', 'weather 2', 'weather 3', 'weather 4']
x=new_train_data[cols]
y=np.log1p(new_train_data['count'])
new_testdata=all_data.iloc[10886:][cols]
```

```python
all_data.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=39)
```

## 采用随机森林算法

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rfr=RandomForestRegressor(n_estimators=300,random_state=40,max_features='log2',oob_score=True)
rfr.fit(x_train, y_train)
rfrpre=rfr.predict(x_test)
print('在随机森林林模型训练集上的均⽅方误差得分: %f'% mean_squared_error(y_test, rfrpre))
```
